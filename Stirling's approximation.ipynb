# Winning Lottery Twice

Let us consider the "Winning Lottery Twice" problem stated in page 32 of Session 1's slides. 

Suppose we accept that the assumptions stated are reasonably correct. 

Then using the approach outlined in "Birthday Problem", the probability of no repeated winner within any given year is:

> P(no repeated winner) = 1,000,000! / (1,000,000 - 365)! * 1 / 1,000,000^365

And consequently the probability of having at least 1 repeated winner within a year is:

> P(at least 1 repeated winner) = 1 - P(no repeated winner)

We can use Stirling's approximation to calculate this probability.

Here are the workings of using Stirling's approximation to calculate the probabilities given in page 34 of Sesssion 1's slides. Try testing it out in Python if you can get the same values, and apply this method to solve the "Winning Lottery Twice" problem.

# 1) Trying to get the same value as shown on page 34 of Sesssion 1's slides

```python
import math 

def birthdaySolution1(k): 
    p1 = 1 - (math.factorial(365)/math.factorial(365-k)) * (1/365**k) # method 1
    return "{:.1%}".format(p1)

def birthdaySolution2(k): 
    p2 = 1 - (2.71**-k)*(1-k/365)**(k-365+1/2) # method 2
    return "{:.1%}".format(p2)

kList = [1,5,10,20,23,30,40,50]
for k in kList:
    print(k, birthdaySolution1(k)) 

print("----------------------")

for k in kList:    
    print(k, birthdaySolution2(k)) 
```

# 2) Solving the "Winning Lottery Twice" problem

```python
def lotterySolution1(k): 
    p1 = 1 - (math.factorial(1000000)/math.factorial(1000000-k)) * (1/1000000**k) # method 1
    return "{:.1%}".format(p1)

print(lotterySolution1(365))
```

```python
def lotterySolution2(k): 
    p2 = 1 - (2.71**-k)*(1-k/1000000)**(k-1000000+1/2) # method 2
    return "{:.1%}".format(p2)

print(lotterySolution2(365))
```

# 3) Using For Loop to solve it

```python
#For 1 Year
probabilityOfNoRepeatWinnerPerDay = 1
totalProbabilityOfNoRepeatWinner = 1
numberOfDays = 365
totalNumberOfPeople = 1000000

def probabilityDifferentWinnerEveryday():
    global probabilityOfNoRepeatWinnerPerDay, totalProbabilityOfNoRepeatWinner
    for a in range (1,numberOfDays):
        totalProbabilityOfNoRepeatWinner = totalProbabilityOfNoRepeatWinner * probabilityOfNoRepeatWinnerPerDay;
        probabilityOfNoRepeatWinnerPerDay = probabilityOfNoRepeatWinnerPerDay - (1/totalNumberOfPeople);    
    return totalProbabilityOfNoRepeatWinner
        
noWinnerProbability = probabilityDifferentWinnerEveryday()

#Probability of having at least 1 repeated winner within a year is:
winnerProbability = 1 - noWinnerProbability

print("{:.2%}".format(winnerProbability))
```
