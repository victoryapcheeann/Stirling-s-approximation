Then using the approach outlined in "Birthday Problem", the probability of no repeated winner within any given year is:

> P(no repeated winner) = 1,000,000! / (1,000,000 - 365)! * 1 / 1,000,000^365

And consequently the probability of having at least 1 repeated winner within a year is:

> P(at least 1 repeated winner) = 1 - P(no repeated winner)
